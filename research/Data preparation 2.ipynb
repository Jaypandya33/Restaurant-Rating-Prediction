```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("cleaned_zomato.csv")
df
```

```python
df.isnull().sum()
```

```python
df['rate'] = df['rate'].str.replace('/5', '', regex=False)
df.head()
```

```python
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')
```

```python
mean_rate = df['rate'].mean()

mean_rate
```

```python
df['rate'] = df['rate'].fillna(mean_rate)
df.head()
```

```python
df.sample()
```

```python
df['approx_cost(for two people)'] = pd.to_numeric(df['approx_cost(for two people)'], errors='coerce')
```

```python
mean_rate = df['approx_cost(for two people)'].mean()

mean_rate
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].fillna(mean_rate)
df.head()
```

```python
df['rest_type'] = df['rest_type'].fillna('Others')
```

```python
df[df['rest_type'] == 'Others'].head()
```

```python
df['cuisines'] = df['cuisines'].fillna('Others')
```

```python
df[df['cuisines'] == 'Others'].head()
```

```python
df['location'] = df['location'].fillna('Others')
```

```python
df[df['location'] == 'Others'].head()
```
